```python
titanic.head(2)
```

```python
pd.get_dummies(titanic['Sex'].head())
```

```python
sex=pd.get_dummies(titanic['Sex'],drop_first=True)
sex.head(3)
```

```python
embark=pd.get_dummies(titanic['Embarked'])
```

```python
embark.head()
```

```python
embark=pd.get_dummies(titanic['Embarked'], drop_first=True)
```

```python
embark.head(3)
```

```python
Pcl=pd.get_dummies(titanic['Pclass'], drop_first=True)
Pcl.head(3)
```

```python
titanic=pd.concat([titanic, sex, embark,Pcl], axis=1)
```

```python
titanic.head(3)
```

```python
titanic.drop(['Name' , 'PassengerId' , 'Pclass' , 'Ticket' , 'Sex' , 'Embarked'], axis=1, inplace=True)
```

```python
titanic.head(3)
```

```python
X=titanic[["Age", "SibSp", "Parch", "Fare"]]
y=titanic['Survived']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.33, random_state=4)
```

```python
lm=linear_model.LogisticRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
prediction= lm.predict(X_test)
```

```python
print("Mean Square Error", mean_squared_error(y_test,prediction))
```

```python
plt.plot(X_test,y_test)
plt.xlabel("Features")
plt.ylabel("Chances of Survived")
```
